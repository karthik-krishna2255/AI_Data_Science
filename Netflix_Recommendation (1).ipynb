```python
df = df.reset_index().rename(columns={'index': 'UserID'})
```

## Data Cleaning

```python
df.dropna(inplace=True)
df['rating'] = df['rating'].astype(float)
```

```python
df.info()
```

## Most Popular Genre

```python
plt.figure(figsize=(12,6))
popular_genre = df['rating'].value_counts()
popular_genre.plot(kind='bar', title='Most Popular Genres')
plt.show()
```

## Most Liked Genre

```python
plt.figure(figsize=(12,6))
liked_genre = df.groupby('certificate')['rating'].mean().sort_values(ascending=False)
liked_genre.plot(kind='bar', title='Average Rating per Votes')
plt.xticks(rotation=45)
plt.show()
```

## Best Movie Recommendation per Genre for a User

```python
df.info()
```

```python
user_id = 10
user_data = df[df['UserID'] == user_id]
recommendations = {}

for genre in df['genre'].unique():
    genre_movies = df[df['genre'] == genre]
    watched_movies = user_data['title'].tolist()
    unwatched = genre_movies[~genre_movies['title'].isin(watched_movies)]
    best_movie = unwatched.groupby('title')['rating'].mean().sort_values(ascending=False).head(1)
    if not best_movie.empty:
        recommendations[genre] = best_movie.index[0]

recommendations
```

## Best and Worst Rated Genres

```python
genre_ratings = df.groupby('genre')['rating'].mean().sort_values(ascending=False)
best_genre = genre_ratings.idxmax()
worst_genre = genre_ratings.idxmin()
print(f"Best_Genre:{best_genre}")
print(f"Worst_Genre:{worst_genre}")
```
